### Baseline Performance

```python
X_train_base, X_test_base, y_train_base, y_test_base, X_val_base, y_val_base = baseline.set_train_test_val_data_by_index(data_with_nulls.X_train.index, data_with_nulls.X_test.index, data_with_nulls.X_val.index)
X_train_base.shape, X_test_base.shape, X_val_base.shape
```

```python
best_base_model = DecisionTreeClassifier(criterion='entropy', max_depth=10, max_features=0.6)
encoder = ColumnTransformer(transformers=[
                ('categorical_features', OneHotEncoder(categories=[list(set(baseline.X_train[col])) for col in baseline.categorical_columns], sparse=False), baseline.categorical_columns),
                ('numerical_features', StandardScaler(), baseline.numerical_columns)])
base_model = Pipeline([
                            ('features', encoder),
                            ('learner', best_base_model)
                        ])
```

```python
n_estimators = 200
boostrap_size = 0.7

ensemble, ensemble_preds = baseline.fit_ensemble_with_bootstrap(base_model, n_estimators, boostrap_size, True)
```

```python
pred_df = pd.DataFrame(ensemble_preds, index=baseline.X_test.index).transpose()
pred_df
```

```python
metrics={}
for group_name in baseline.groups.keys():
    temp = pred_df[baseline.groups[group_name].index]
    res = {}
    res['std'] = temp.std().values
    res['iqr'] = sp.stats.iqr(temp, axis=0)

    y_preds = np.array([int(x<0.5) for x in temp.mean().values])
    res['label_stability']=compute_label_stability(y_preds)
    metrics[group_name] = res
```

```python
for g in baseline.groups.keys():
    print (g)
    print (baseline.y_test[baseline.groups[g].index].value_counts())
```

```python
std_to_plot = [metrics[g]['std'] for g in baseline.groups.keys()]

plt.figure(figsize=(10,6))
plt.boxplot(std_to_plot, vert=True, patch_artist=False, labels=baseline.groups.keys()) 
plt.ylabel('Standard deviation')
plt.title('Baseline: Std of predicted probabilities')  
```

```python
iqr_to_plot = [metrics[g]['iqr'] for g in baseline.groups.keys()]

plt.figure(figsize=(10,6))
plt.boxplot(iqr_to_plot, vert=True, patch_artist=False, labels=baseline.groups.keys()) 
plt.ylabel('Inter quantile range')
plt.title('Baseline: IQR of predicted probabilities')  
```

```python
for group_name in baseline.groups.keys():
    X_test_group = baseline.groups[group_name]
    res_df = pd.DataFrame({})
    for m in range(n_estimators):
        y_preds = ensemble[m].predict(X_test_group)
        cf = confusion_matrix(baseline.y_test[X_test_group.index], y_preds)
        res_df = res_df.append(pd.DataFrame(confusion_matrix_metrics(cf),index=[m]))
    metrics[group_name]['cf_metrics'] = res_df        
        
```

```python
metric='TPR'
to_plot = [metrics[g]['cf_metrics'][metric] for g in baseline.groups.keys()]

plt.figure(figsize=(10,4))
plt.boxplot(to_plot, vert=True, patch_artist=False, labels=baseline.groups.keys()) 
plt.ylabel(metric)
plt.title('Baseline:'+metric)  
plt.show()
```
